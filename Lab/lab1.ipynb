```python
## Write a program to check if a given string is palindrome 

str=input("Enter the string").lower()
r_str=str[::-1]

def palindrome_(str,r_str):
    if str==r_str:
        print(f"The given {str} is palindrome")
    else:
        print("Not")

palindrome_(str,r_str)
    
    
    
```

```python
# prints all the prime numbers upto n 

n=int(input("Enter the number upto which you want to print prime numbers"))

def fact_(n):
    for i in range(2,n+1):
        prime=0
        for j in range (1,i+1):
            if i%j==0:
                prime+=1
        if prime==2:
            print(i)

fact_(n)

    
```

```python
#that will print fibonacci series upto n term 

n=int(input("Enter the valid number upto which you want to print  fibonacci series"))

def fib_(n):
    if n<=1:
        return n 
    else:
        return fib_(n-1)+fib_(n-2)


i=1
while True:
    val=fib_(i)
    if val>n:
        break
    print(val)
    i+=1

```

```python
# Define a function that takes two numbers as arguments, and returns their greatest common divisor
n1=int(input("Enter the number"))
n2=int(input("Enter the number"))

def gcd_(n1,n2):
    gcd=1
    for i in range(1,min(n1,n2)+1):
        if n1%i==0 and n2%i==0:
            gcd=i

    return gcd
print(gcd_(n1,n2))
```
