### list slicing 

```python
a=[1,2,3,4,5,6,7,8,9]
print(a[1:4])
print(a[:3:3])
print(a[-2:])


b=['aa',1,4.5,7,9,4,3]
print(b[::-1])
print(b[2:-2:2])
```

## TUPLE

```python
#why tuple is faster than lsit 

import time as t 

start = t.time()
for _ in range(5_000_000):
    a=[1,2,3,4]
end=t.time()
print("list exec time:",end-start)

start = t.time()
for _ in range(5_000_000):
    a=(1,2,3,4)
end=t.time()
print("Tuple exec time:",end-start)
```
