```python
#take 3 number and ask for the largest number 

num1=int(input("Enter the 1 integer:"))
num2=int(input("Enter the 2 integer:"))
num3=int(input("Enter the 3 integer:"))

def large(num1,num2,num3):
    if num1>num2 and num1>num3:
        print(f"{num1} is the largest")
    elif num2>num3:
        print(f"{num2} is the largest")
    else:
        print(f"{num3} is the largest")
    

large(num1,num2,num3)



        
```

```python
#multiplication of number and highlighting the multiple of 5 

n=int(input("Enter an integer:\n"))

def mul(n):
    for i in range(1,11):
        r=i*n
        if r%5==0:
            print(f"\033[1m{n}*{i}={r}\033[0m")
        else:
            print(f"{n}*{i}={r}")

       
        
mul(n)

```
